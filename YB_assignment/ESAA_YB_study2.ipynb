## 1-4. 데이터 핸들링 - 판다스

### 판다스 시작 - 파일을 DataFrame으로 로딩, 기본 API  
read_csv('filepath', sep = ',', ...): 기본 필드 구분 문자는 콤마(,), 탭일 경우 '\t'  
DataFrame 객체의 shape 변수: 행과 열 수를 튜플 형태로 반환  
info() 메서드: 총 데이터 건수, 데이터 타입, Null 건수 등의 정보  
describe() 메서드: 숫자형 칼럼의 분포도  
value_counts() 메서드: Series 객체에 호출 시 해당 칼럼값의 데이터값 건수 반환, 기본적으로 Null값을 제외하고 건수를 계산, 포함하려면 dropna=False 옵션 사용  
모든 Series와 DataFrame은 인덱스를 반드시 가짐.  

### DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환  
DataFrame을 ndarray로 변환: DataFrame.values  
DataFrame을 리스트로 변환: DataFrame.values.tolist()  
DataFrame을 딕셔너리로 변환: DataFrame.to_dict('list')

### DataFrame 데이터 삭제  
DataFrame.drop(lables=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')  
-axis=1 입력 시 labels 값을 칼럼명으로, axis=0 입력 시 labels 값을 인덱스로 간주  
-inplace=False이면 원래 DataFrame의 데이터는 삭제하지 않고 삭제한 결과 DataFrame을 반환, inplace=True로 설정하면 원래 DataFrame의 데이터를 삭제하고 None을 반환

### Index 객체  
DataFrame.index 또는 Series.index 속성을 통해 Index 객체 추출  
Index 객체는 식별성 데이터를 1차원 array로 가지고 있음.  
한 번 만들어진 DataFrame 및 Series의 Index 객체는 함부로 변경 불가.  
reset_index() 메서드: 인덱스 새로 할당, 기존 인덱스는 'index'라는 칼럼으로 추가.  
-Series에 적용시 DataFrame이 반환됨.  
-drop=True 설정 시 기존 인덱스 삭제

### 데이터 셀렉션 및 필터링  
DataFrame[]: 칼럼명 또는 불린 인덱스를 입력  
DataFrame.iloc[]: 위치 기반 인덱스, 위치 정숫값 및 위치 정수 슬라이싱 입력, 불린 인덱싱X  
DataFrame.loc[]: 명칭 기반 추출, 인덱스값과 칼럼명 입력, 슬라이싱 적용 시 종료 값까지 포함

### 정렬, Aggregation 함수, GroupBy 적용  
DataFrame.sort_values(by=[], ascending=True, inplace=False)  
DataFrame.groupby(by='')  
groupby에 aggregation 함수를 적용할 때  
-함수 하나이면 바로 적용  
-여러 개면 agg() 내에 인자로 입력, 칼럼마다 다른 함수를 호출하려면 딕셔너리 형태로 입력

### 결손 데이터 처리하기  
DataFrame.isna(): 결손 데이터 여부를 True/False로 반환, sum()을 붙여 개수 확인  
DataFrame.fillna(''): 결손 데이터 대체, 반환 값을 다시 받거나 inplace=True 파라미터를 추가해야 적용

### apply lambda 식으로 데이터 가공  
DataFrame.apply(lambda 식)  
-if문에서 elif는 지원 X  
-식이 복잡해질 경우 def로 함수를 따로 만들어 lambda로 적용하는 것이 나음

### apply lambda 식으로 데이터 가공

```python
def get_square(a):
  return a**2

print('3의 제곱은:', get_square(3))
```

```python
lambda_square = lambda x : x ** 2
print('3의 제곱은:', lambda_square(3))
```

```python
a = [1, 2, 3]
squares = map(lambda x : x**2, a)
list(squares)
```

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']].head(3)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(8)
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x <= 60
                                                                                  else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
def get_category(age):
  cat = ''
  if age <= 5: cat = 'Baby'
  elif age <= 12: cat = 'Child'
  elif age <= 18: cat = 'Teenager'
  elif age <= 25: cat = 'Student'
  elif age <= 35: cat = 'Young Adult'
  elif age <= 60: cat = 'Adult'
  else: cat = 'Elderly'

  return cat

# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정.
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
